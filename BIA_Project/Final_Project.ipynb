```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd
```

```python
def getReviews(page_url):
    companies = []
    ratings=[]
    fields=[]
    texts=[]
    page = requests.get(page_url)
    if page.status_code==200:
        soup = BeautifulSoup(page.content, 'html.parser')
    for i in soup.find_all('a', class_ = 'css-1s5eo7v eu4oa1w0'):
        company = i.get_text()
        companies.append(company)
#     for j in soup.find_all('div', class_ = 'css-kbqft4 e1wnkr790'):
#         field = i.get_text()
#         fields.append(field)
        
    for n in soup.find_all('div',class_='css-11ksq3 eu4oa1w0'):
        if n.find('span',class_='css-bvu7jx e1wnkr790'):
            star=n.find('span',class_='css-bvu7jx e1wnkr790').text
            ratings.append(star)
        else:
            ratings.append("None")
            
    for m in soup.find_all('div',class_='css-11ksq3 eu4oa1w0'):
        if m.find('div',class_='css-kbqft4 e1wnkr790'):
            field=m.find('div',class_='css-kbqft4 e1wnkr790').text
            fields.append(field)
        else:
            fields.append("None")
    
    
    for s in soup.find_all('div',class_='css-tzqc93 e37uo190'):
        if s.find('p',class_='css-h215jb e1wnkr790'):
            textz=s.find('p',class_='css-h215jb e1wnkr790').text
            texts.append(textz)
        else:
            texts.append("None")
            
            
    df = pd.DataFrame({'companies' : companies[1:],'Ratings':ratings[1:],"Fields":fields[1:],"Description":texts[1:]})
    return df
```

```python
# enter your own url
page_url = "https://www.indeed.com/companies/search?q=software+developer&l=new+york"

reviews=getReviews(page_url)
reviews
```
