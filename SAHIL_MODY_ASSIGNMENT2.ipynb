- Q1. Write a function to scrape all **reviews on the first page**, including, 
    - **reviewer's name** (see (1) in Figure)

    - **review content** (see (4) in Figure. For each review text, need to get the **complete text**, i.e., need to expand the `more` button. Only text is needed, pictures are not needed. Hint: take a close look at the content of the html file.)
    - **likes** (see (5) in Figure). 
    - If a field, e.g. score, is missing, use `None` to indicate it. 
- `Function Input`: book page URL
- `Function Output`: save all reviews as a DataFrame of columns (`reviewer, rating, date, review, like`). E.g., for the given URL, you can get 30 reviews.(10 points)
- **Note**: GoodReads occasionaly blocks request. You may get an error that is not due to your codes. You may try a couple of times.

```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd
 

def getReviews(page_url):
    page=requests.get(page_url)
    rows=[]
    authors=[]
    date=[]
    likes=[]
    review=[]
    rating=[]
    if page.status_code==200:
        soup=BeautifulSoup(page.content,'html.parser')
    for i in soup.find_all("a",class_="user"):
        a=i.get_text()
        authors.append(a)

    for j in soup.find_all("a",class_="reviewDate createdAt right"):
        b=j.get_text()
        date.append(b)
        
    for k in soup.find_all("div",class_="updateActionLinks"):
        c=k.find("span",class_="likesCount")
        d=c.get_text()
        likes.append(d.strip("likes"))
        
    for m in soup.find_all("div",class_='reviewText stacked'):
        e=m.get_text()
        review.append(e.strip("\n"))
        
    for n in soup.find_all('div',class_='reviewHeader uitext stacked'):
        if n.find('span',class_='staticStars notranslate'):
            rates=n.find('span',class_='staticStars notranslate').text
            rating.append(rates)
        else:
            rating.append("None")
        
    df=pd.DataFrame({"name":authors,"rating":rating,"date":date,"review":review,"like":likes})
    return df
    
```
